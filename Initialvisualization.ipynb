```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde
```

```python
file = '/content/drive/MyDrive/uRAD_PSD.csv'
data = pd.read_csv(file)

total = data.iloc[:,0].values
head = data.iloc[:,1].values

print(total)
print(head)
```

```python
head = head*100+1
total = total*100+1
ratio = (total - head) / total
x = total
y = ratio

print(x,y)
```

```python
k = gaussian_kde([x, y])
density = k.evaluate([x, y])

plt.scatter(x, y, c=density, cmap='jet', alpha=0.5)

plt.ylim(0, 0.6)

plt.show()
```
